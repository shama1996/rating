```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import urllib
from selenium.webdriver.remote.webelement import WebElement
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome("C:\Program Files (x86)\chromedriver_win32\latest\chromedriver.exe")
```

```python
def scrape_(url,category):

    final_list=[]
    final_review_list=[]
    itagCount=0
    pageCount=0
    pageCount=0
    final_list=[]
    wait = WebDriverWait(driver, 30);
    

    while(pageCount<500):
        if pageCount>0:
            try:
                driver.get(url+'-page-'+str(pageCount+1))
                wait.until(expected_conditions.visibility_of_element_located((By.XPATH,'//div[@class="col-10 review"]/div[@class="more reviewdata"]')))     
            except Exception as e:
                print('page reached to end')
                break
    
        else:
            driver.get(url)
            
        pageCount=pageCount+1;    
        print('page count',pageCount) 
        print('current url',driver.current_url) 
        wait.until(expected_conditions.visibility_of_element_located((By.XPATH,'//div[@class="col-10 review"]/div[@class="more reviewdata"]')))       
        review_list=driver.find_elements_by_xpath('//div[@class="col-10 review"]/div[@class="more reviewdata"]')
        rating_list=driver.find_elements_by_xpath('//div[@class="col-10 review"]/div[@class="rating"]/span[1]')

        length=len(review_list)
        print('rating list >>',len(rating_list))
        print('review list >>',len(review_list))
    
        for i in range(0,length):
            review=''
            review=review_list[i].text

            #find rating star.Here if class name is 'icon-rating rated-star' then need to consider those i tag for star
            itagCount=0
            itag_list=rating_list[i].find_elements_by_xpath('./child::*')

            for itag in itag_list:
                if itag.get_attribute("class")=='icon-rating rated-star':
                    #print(itag.get_attribute("class"))
                    itagCount=itagCount+1

            final_list.append((category,review,str(itagCount)))
        
        
    print('final data list size>>',len(final_list))
    return final_list
```
